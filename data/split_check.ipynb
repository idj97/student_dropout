```python
import pandas as pd
from sklearn.model_selection import train_test_split

train_size = 0.6
validation_size = 0.2 
test_size = 0.2
assert(train_size + validation_size + test_size == 1)

df = pd.read_csv("data.csv")
train, holdout = train_test_split(
    df, 
    test_size=validation_size+test_size, 
    train_size=train_size,
    random_state=0,
    shuffle=True
)

n = int(len(df)*validation_size)
validation = holdout[:n]
test = holdout[n:]
len(train), len(validation), len(test)
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

for field in validation.columns:
    try:
        sb.kdeplot(data=validation, x=field)
        sb.kdeplot(data=test, x=field)
        plt.show()
    except:
        print("failed to show", field)
        pass
```
